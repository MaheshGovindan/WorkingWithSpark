```python
from pyspark.sql import SparkSession

spark= SparkSession.builder.appName("LearningSparkJul2025").getOrCreate()
```

```python
data1 = [("A",1001),("B",1002),("C",1003),("D",1004),("E",1005)]
df1=spark.createDataFrame(data1,["key","value1"])
df1
```

```python
data2=[("B",10000),("E",10002),("F",10003)]
df2=spark.createDataFrame(data2,["key","value2"])
df2
```

```python
print("Inner Join")
df1.join(df2,"key","inner").show()
```

```python
print("Full Outer Join")
df1.join(df2,"key","full_outer").show()
```

```python
print("Left Outer Join")
df1.join(df2,"key","left_outer").show()
```

```python
print("Right Outer Join")
df1.join(df2,"key","right_outer").show()
```

```python
print("Left Semi Join")
df1.join(df2,"key","left_semi").show()
```

```python
print("Left Anti Join")
df1.join(df2,"key","left_anti").show()
```

```python
print("Cross Join")
df1.join(df2,how="cross").show()
```
